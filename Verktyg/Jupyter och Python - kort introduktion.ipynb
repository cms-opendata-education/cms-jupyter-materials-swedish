### 2. Skapa grafer

Grafer är ett ytterst viktigt verktyg för undervisning i naturvetenskapliga ämnen, men det tar ofta onödigt länge att rita dem i en undervisningssituation. Med hjälp av Jupyter går det snabbt och smidigt att skapa grafer, antingen genom att mata in värden själv, eller genom att läsa in en datafil med värden. Datafilen i fråga kan innehålla tiotusentals värden utan att det skapar minsta problem, vilket gör programmering till ett oumbärligt verktyg för att tolka statistik.

```python
siffror = [1,2,3,6,7,8,12,16]   # Vi definierar variabeln 'siffror': En lista med siffervärden.
plt.plot(siffror, 'b*')         # plt.plot skapar en graf, b* anger markörerna: blå stjärnor.
plt.show()
```

```python
# Vi definierar en ny variabel - en lista med åldrar. Variabler i python kan inte innehålla å, ä eller ö.
# De kan inte börja med en siffra (ex. 1data, 2data), och är skriftlägeskänsliga/case sensitive.

alder = [14,15,14,16,17,16,14,15,16,16,14,15,15,12,34,12,15,16,16]

# Vi ritar ett histogram med funktionen plt.hist.
# Parametern bins= anger antalet staplar, och range= anger histogramets spann. 

plt.hist(alder, bins=31, range=(10,40))

# För att göra diagramet tydligare sätter vi in namn på koordinataxlarna och en rubruk för diagramet:

plt.xlabel('Åldrar')
plt.ylabel('Antal')
plt.title('Gruppens åldersfördelning \n') # \n skapar en ny rad för att separera rubriken från grafen.

# Slutligen ber vi programmet visa grafen:

plt.show()
```

### 3. Att skapa grafer

Här läser vi in en datafil att undersöka. Denna datafil beskriver solens aktivitet från 1992 till nutid. I datafilen medföljer ingen förklaring, men [här](http://sidc.oma.be/silso/infosndhem) ser vi vad varje kolumn betyder. Vi undersöker solfläckarnas antal: I kolumn 4 finns antalet solfläckar för varje dag, och i kolumn 5 och 6 är de indelade efter solens norra respektive södra halvklot.

Vi definierar variabeln *soldata*. Funktionen read_table som hör till paketet pandas kan läsa in en csv-fil. csv står för comma separated values, och om vi öppnar filen i ett textprogram ser den ut så här:

<img src = ../Kuvat/soldatacsv.png>


Vanligtvis är värden i en csv-fil separerade med kommatecken, men i det här fallet används semikolon istället. Vi måste beakta detta när vi läser in den.

```python
soldata = pd.read_table('http://sidc.oma.be/silso/INFO/sndhemcsv.php', sep=';', encoding = "ISO-8859-1", header = None)

# pd.read_table läser in en csv-fil och gör den till en variabel i dataframe-format (en tabell)
# Vi anger filens sökväg. Om filen skulle finnas på din dator i samma mapp som detta dokument
# kunde du också bara hänvisa till den med dess filnamn. 

# sep=';' gör så att programmet tolkar semikolon som separatorer istället
# encoding = ändrar på programmets textpaket, och kan behövas för filer som innehåller specialtecken.
# utan parametern header = None skull programmet anta att den första raden i tabellen innehåller kolumnernas rubriker.


```

```python
# Med kommandot .head kan vi visa de första raderna av tabellen och se att allt ser ok ut.

soldata.head()
```

```python
# Vi skapar en tid-antal-graf.

fig = plt.figure(figsize=(15, 4))
plt.scatter(soldata[3],soldata[4], s=0.2)  

# plt.scatter skapar en scatter plot - ett diagram där varje talpar bildar en punkt.
# De första parametrarna anger listorna för talpar
# soldata[3] är kolumnen '3' i tabellen - innehållande "datum" i form av decimaltal.
# soldata[4] innehåller antalet solfläckar under den dagen
# parametern s= anger markörernas storlek.

plt.title("Solfläckarnas aktivitet \n")
plt.ylabel('Dagligt antal solfläckar')
plt.xlabel('Årtal')

plt.show()
```

```python
# soldata[5] och soldata[6] innehåller norra och södra halvklotets solfläckar.
# Vi ber programmet rita diagram över båda, så kommer de att visas i samma graf.

fig = plt.figure(figsize=(15, 4))
plt.scatter(soldata[3],soldata[5], s=0.5, label='Norra halvklotet', alpha=0.5)
plt.scatter(soldata[3],soldata[6], s=0.5, label='Södra halvklotet', alpha=0.5)

# label= namnger graferna, och behövs för att .legend-kommandot nedan ska fungera
# alpha= kan användas för att ställa ner färgstyrkan.


plt.title("Solfläckarnas aktivitet \n")
plt.ylabel('Dagligt antal solfläckar')
plt.xlabel('Årtal')

plt.legend(loc='upper right')  # .legend skapar en referensruta

plt.show()
```
